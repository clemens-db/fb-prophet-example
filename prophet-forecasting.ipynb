```python
# Python
import pandas as pd
import numpy as np
from fbprophet import Prophet
%matplotlib inline
```

```python
# Python
df = pd.read_csv('./data/example_wp_peyton_manning.csv')
df.head()
```

```python
# Python
def forecast(tdf):
  forecast_df = tdf
  
  m = Prophet()
  m.fit(forecast_df)
  
  future = m.make_future_dataframe(periods=365)
  forecast = m.predict(future)
  
  m.plot(forecast)
  
  return m
```

```python
forecast(df)
```

```python
# Python
m.plot_components(forecast);
```
